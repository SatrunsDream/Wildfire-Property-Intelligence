Loads the compressed dataset, inspects variable names and types, confirms the unit of analysis

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
DATA_PATH = Path('../../dataset/Capstone2025_nsi_lvl9_with_landcover_and_color.csv.gz')
df = pd.read_csv(DATA_PATH, compression='gzip', low_memory=False)
print(f'Dataset shape: {df.shape}')
print(f'\nColumns: {list(df.columns)}')
```

```python
print('\nData types:')
print(df.dtypes)
print('\nFirst few rows:')
display(df.head())
```

```python
print('Unique H3 cells:', df['h3'].nunique())
print('Unique counties (FIPS):', df['fips'].nunique())
print('Duplicates at H3 x landcover:', df.duplicated(subset=['h3', 'lc_type']).sum())
print('\nLandcover frequency:')
display(df['lc_type'].value_counts().rename_axis('lc_type').reset_index(name='count'))
```
